# 5) How many Movies & TV Shows are in the dataset ? Show with Bar Graph.

```python
netflix['Category'].value_counts()
```

```python
category_counts = netflix['Category'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
category_counts.plot(kind='bar', color=['blue', 'green'])
plt.title("Count of Movies and TV Shows")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=0)
for index, value in enumerate(category_counts):
    plt.text(index, value + 50, str(value), ha='center', fontsize=10)
plt.show()
```
